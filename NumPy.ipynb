```python
import numpy as np
```

Python lists are limited in functions and take up more space and time to process than NumPy arrays.

NumPy provides an array object that is much faster and more compact than Python lists.

The array object in NumPy is called `ndraary` meaning 'n-dimensional array'.

```python
one_dimensional_arr = np.array([10, 12])
print(one_dimensional_arr)
```

There are several ways to create an array in NumPy. We can create a 1-D array by simply using the function `array()` which takes in a list of values as an argument and returns a 1-D array.

```python
# Create and print a NumPy array 'a' containing the elements 1, 2, 3.
a = np.array([1, 2, 3])
print(a)
```

Another way to implement an array is using `np.arange()`. This function will return an array of evenly spaced values within a given interval.

```python
# Create an array with 3 integers, starting from the default integer 0.
b = np.arange(3)
print(b)
```

```python
#Create an array that starts from the integer 1, ends at 20, incremented by 3.
c = np.arange(1, 20, 3)

# np.arange(start, stop, step)
print(c)
```

```python
d = np.arange(0, 100, 5)
print(d)
```

```python
# np.linspace(start, stop, num)
# step = (stop - start) / (num - 1)
a = np.linspace(0, 100, 21)
print(a)
```

```python
lin_spaced_arr_int = np.linspace(0, 100, 5, dtype=int)
print(lin_spaced_arr_int)
```

In np.arange stop is not included but in np.linspace stop is included. np.arange is in int but in np.linspace is in float

```python
b_float = np.arange(3, dtype=float)
print(b_float)
```

# More on NumPy Arrays
One of the advantages of using NumPy is that we can easily create arrays with built-in functions such as:
* `np.ones()` - Returns a new array setting values to one.
* `np.zeros()` - Returns a new array setting values to zero.
* `np.empty()` - Returns a new uninitialized array.
* `np.random.rand()` - Returns a new array with values chosen at random.
